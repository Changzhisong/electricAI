```python
import pandas as pd
import numpy as np

from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_regression
```

```python
months = range(1, 10)
```

```python
feture_path_fmt = '../dataset/fetures/{}/train_feture.csv'
```

```python
fetures = []
```

```python
for month in months:
    feture_path = feture_path_fmt.format(month)
    df = pd.read_csv(feture_path)
    df.info()
    fetures.append(df)
```

```python
combine_fetures = pd.concat(fetures)
combine_fetures.info()
```

## 删掉为空的特征

```python
combine_fetures.shape
```

```python
# combine_fetures.info(verbose=True, null_counts=True)
```

```python
# combine_fetures.dropna(axis=1, how='all').info(verbose=True, null_counts=True)
```

```python
combine_fetures = combine_fetures.dropna(axis=1, how='all')
```

```python
combine_fetures.info()
```

```python
X = combine_fetures.drop(['predict_date', 'predict_power_consumption'], axis=1)
```

```python
y = combine_fetures['predict_power_consumption']
```

```python
select = SelectKBest(f_regression, k=3000)
```

```python
new_x = select.fit_transform(X.fillna(0), y)
```

```python
use_index = select.get_support()
```

```python
use_columns = [x for x in X.columns[use_index]] + ['predict_date', 'predict_power_consumption']
```

```python
columns = [x for x in use_columns if x not in ['predict_date']]
feture_date = combine_fetures['predict_date']
```

```python
combine_fetures = combine_fetures[columns].apply(lambda x:x.fillna(0), axis=1)
combine_fetures['predict_date'] = feture_date
combine_fetures.info(verbose=True, null_counts=True)
```

```python
combine_fetures.to_csv('../dataset/fetures/train_feture.csv', index=False)
```

```python
use_columns = combine_fetures.columns
```

# 删除预测值的特征

```python
test_feture = pd.read_csv('../dataset/fetures/test_feture.csv')
```

```python
test_feture = test_feture[use_columns]
```
